WE HAVE ALREADY EXAMINED THE CLASS DISTRIBUTION AND OBSERVED THAT SOME FEATURES ARE HIGHLY CORRELATED WITH EACH OTHER. THIS INDICATES A POTENTIAL PRESENCE OF MULTICOLLINEARITY. THEREFORE, WE ARE PROCEEDING TO PERFORM VARIANCE INFLATION FACTOR (VIF) ANALYSIS TO DETECT AND ADDRESS MULTICOLLINEARITY.

```python
import pandas as pd
from statsmodels.stats.outliers_influence import variance_inflation_factor

df = pd.read_csv('C:/Users/Admin/Downloads/BankChurners.csv')
df.drop(columns=['CLIENTNUM', 'Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_1', 
                 'Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_2'], inplace=True)

df['Card_Category'] = df['Card_Category'].replace(['Gold', 'Platinum'], 'Others')

df = pd.get_dummies(df, columns=['Education_Level', 'Income_Category', 'Marital_Status', 'Card_Category'], dtype=int, drop_first=True)
LB = preprocessing.LabelBinarizer()
df['Gender'] = LB.fit_transform(df['Gender'])
df['Attrition_Flag'] = LB.fit_transform(df['Attrition_Flag'])

print('##########  VIF Analysis  ##########\n\n')
independent_variables = ['Customer_Age', 'Gender', 'Dependent_count', 'Months_on_book', 'Total_Relationship_Count', 
                         'Months_Inactive_12_mon', 'Contacts_Count_12_mon', 'Credit_Limit', 'Total_Revolving_Bal', 'Avg_Open_To_Buy', 'Total_Amt_Chng_Q4_Q1', 
                         'Total_Trans_Amt', 'Total_Trans_Ct', 'Total_Ct_Chng_Q4_Q1', 'Avg_Utilization_Ratio', 'Education_Level_Doctorate', 'Education_Level_Graduate', 
                         'Education_Level_High School', 'Education_Level_Post-Graduate', 'Education_Level_Uneducated', 'Education_Level_Unknown', 
                         'Income_Category_$40K - $60K', 'Income_Category_$60K - $80K', 'Income_Category_$80K - $120K', 'Income_Category_Less than $40K', 
                         'Income_Category_Unknown', 'Marital_Status_Married', 'Marital_Status_Single', 'Marital_Status_Unknown', 'Card_Category_Others', 
                         'Card_Category_Silver']

while True:
    x = df[independent_variables]
    vif = [variance_inflation_factor(x.values, i) for i in range(x.shape[1])]
    maxloc = vif.index(max(vif))
    if max(vif) > 10:
        print('VIF: ', vif)
        print('Dropping', independent_variables[maxloc], 'in index:', str(maxloc), '\n')
        del independent_variables[maxloc]
    else:
        break

print('independent_variables:', independent_variables)
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import sklearn.linear_model as lm
from sklearn import metrics
import numpy as np

df = pd.read_csv('C:/Users/Admin/Downloads/BankChurners.csv')
df.drop(columns=['CLIENTNUM', 'Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_1', 'Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_2'], inplace=True)

df['Card_Category'] = df['Card_Category'].replace(['Gold', 'Platinum'], 'Others')

df = pd.get_dummies(df, columns=['Education_Level', 'Income_Category', 'Marital_Status', 'Card_Category'], dtype=int, drop_first=True)
LB = preprocessing.LabelBinarizer()
df['Gender'] = LB.fit_transform(df['Gender'])
df['Attrition_Flag'] = LB.fit_transform(df['Attrition_Flag'])

x = df[['Gender', 'Dependent_count', 'Total_Relationship_Count', 'Months_Inactive_12_mon', 'Contacts_Count_12_mon', 'Total_Revolving_Bal', 
        'Avg_Open_To_Buy', 'Total_Trans_Amt', 'Total_Ct_Chng_Q4_Q1', 'Avg_Utilization_Ratio', 'Education_Level_Doctorate', 'Education_Level_Graduate', 
        'Education_Level_High School', 'Education_Level_Post-Graduate', 'Education_Level_Uneducated', 'Education_Level_Unknown', 
        'Income_Category_$40K - $60K', 'Income_Category_$60K - $80K', 'Income_Category_$80K - $120K', 'Income_Category_Less than $40K', 
        'Income_Category_Unknown', 'Marital_Status_Married', 'Marital_Status_Single', 'Marital_Status_Unknown', 'Card_Category_Others', 
        'Card_Category_Silver']]
y = df['Attrition_Flag']

sc = StandardScaler()
x = sc.fit_transform(x)

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=2025)

lr = lm.LogisticRegression()
lr.fit(x_train, y_train)
print('Logistic Regression Accuracy -Test:', lr.score(x_test, y_test))
print('\nROC AUC Accuracy -Test:', metrics.roc_auc_score(y_test, lr.predict_proba(x_test)[:, 1]))
print('\nLR Confusion Matrix -Test:\n', metrics.confusion_matrix(y_test, lr.predict(x_test)))
print('\nLR Classification Report: -Test:\n', metrics.classification_report(y_test, lr.predict(x_test)))
print("""Precision (class 0) = 0.76 means: Of all predicted Attrited Customers, \nonly 76% were actually 
Attrited Customers.
\nRecall (class 0) = 0.45 means: It correctly found 45% of the actual Attrited 
Customers.
\nF1 (class 0) = 0.57 means: The balance is too low. So, the model is better at being correct 
when it says someone will leave, but it misses many actual leavers.\n\n""")

fpr, tpr, threshold = metrics.roc_curve(y_train, lr.predict_proba(x_train)[:, 1])
roc_auc = metrics.auc(fpr, tpr)
print('Area Under The Roc Curve:', roc_auc)

plt.plot(tpr, label='TPR')
plt.plot(1-fpr, label='FPR')
plt.xlabel('1 - False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Find Optimal Threshold')
plt.show()

def find_optimal_threshold(target, predicted):
    fpr, tpr, threshold = metrics.roc_curve(y_train, lr.predict_proba(x_train)[:, 1])
    roc = pd.DataFrame({
        'tf': fpr+tpr-1,
        'threshold': threshold
    })
    roc_t = roc.iloc[(roc.tf-0).abs().argsort()[:1]]
    return list(roc_t['threshold'])

threshold = find_optimal_threshold(y_train, lr.predict_proba(x_train)[:, 1])
y_optimal_pred = np.where(lr.predict_proba(x_test)[:, 1] >= threshold, 1, 0)
print('Optimal threshold Accuracy -Test:', metrics.accuracy_score(y_test, y_optimal_pred))
print('\nOptimal threshold Classification Report:\n', metrics.classification_report(y_test, y_optimal_pred))
print("""We are Still finding imbalance, So maybe we need to provide more example of Attrited Customers, 
then we can see if it learns batter from more examples of Attrited Customers, because we want to
predict who will leave.""")
```

```python
import pandas as pd
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import sklearn.linear_model as lm
from sklearn import metrics
import numpy as np
from imblearn.over_sampling import RandomOverSampler, SMOTE

df = pd.read_csv('C:/Users/Admin/Downloads/BankChurners.csv')
df.drop(columns=['CLIENTNUM', 'Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_1', 'Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_2'], inplace=True)

df['Card_Category'] = df['Card_Category'].replace(['Gold', 'Platinum'], 'Others')

df = pd.get_dummies(df, columns=['Education_Level', 'Income_Category', 'Marital_Status', 'Card_Category'], dtype=int, drop_first=True)
LB = preprocessing.LabelBinarizer()
df['Gender'] = LB.fit_transform(df['Gender'])
df['Attrition_Flag'] = LB.fit_transform(df['Attrition_Flag'])

x = df[['Gender', 'Dependent_count', 'Total_Relationship_Count', 'Months_Inactive_12_mon', 'Contacts_Count_12_mon', 'Total_Revolving_Bal', 
        'Avg_Open_To_Buy', 'Total_Trans_Amt', 'Total_Ct_Chng_Q4_Q1', 'Avg_Utilization_Ratio', 'Education_Level_Doctorate', 'Education_Level_Graduate', 
        'Education_Level_High School', 'Education_Level_Post-Graduate', 'Education_Level_Uneducated', 'Education_Level_Unknown', 
        'Income_Category_$40K - $60K', 'Income_Category_$60K - $80K', 'Income_Category_$80K - $120K', 'Income_Category_Less than $40K', 
        'Income_Category_Unknown', 'Marital_Status_Married', 'Marital_Status_Single', 'Marital_Status_Unknown', 'Card_Category_Others', 
        'Card_Category_Silver']]
y = df['Attrition_Flag']

sc = StandardScaler()
x = sc.fit_transform(x)

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=2025)

ROS = RandomOverSampler(random_state=2025)
SMOTE = SMOTE(random_state=2025)

x_ROS_train, y_ROS_train = ROS.fit_resample(x_train, y_train)
x_SMOTE_train, y_SMOTE_train = SMOTE.fit_resample(x_train, y_train)

lr_ros = lm.LogisticRegression()
lr_ros.fit(x_ROS_train, y_ROS_train)

lr_smote = lm.LogisticRegression()
lr_smote.fit(x_SMOTE_train, y_SMOTE_train)

print('LR ROS Accuracy -Test:', metrics.accuracy_score(y_test, lr_ros.predict(x_test)))
print('\nLR ROS Classification Report -Test:\n', metrics.classification_report(y_test, lr_ros.predict(x_test)))
print('\nLR SMOTE Accuracy -Test:', metrics.accuracy_score(y_test, lr_smote.predict(x_test)))
print('\nLR SMOTE Classification Report -Test:\n', metrics.classification_report(y_test, lr_smote.predict(x_test)))
print("""Both ROS and SMOTE help catch more churners. SMOTE performs slightly better overall, 
but both improve minority class detection.
Nevertheless, let's see the different models also with Cross Validation to get a fair
estimate of which model works best""")
```

```python
import pandas as pd
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, cross_validate, StratifiedKFold
import numpy as np
from sklearn import tree
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import GradientBoostingClassifier, ExtraTreesClassifier, RandomForestClassifier, BaggingClassifier, AdaBoostClassifier
import xgboost as xgb

df = pd.read_csv('C:/Users/Admin/Downloads/BankChurners.csv')
df.drop(columns=['CLIENTNUM', 'Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_1', 'Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_2'], inplace=True)

df['Card_Category'] = df['Card_Category'].replace(['Gold', 'Platinum'], 'Others')

df = pd.get_dummies(df, columns=['Education_Level', 'Income_Category', 'Marital_Status', 'Card_Category'], dtype=int, drop_first=True)
LB = preprocessing.LabelBinarizer()
df['Gender'] = LB.fit_transform(df['Gender'])
df['Attrition_Flag'] = LB.fit_transform(df['Attrition_Flag'])

x = df[['Gender', 'Dependent_count', 'Total_Relationship_Count', 'Months_Inactive_12_mon', 'Contacts_Count_12_mon', 'Total_Revolving_Bal', 
        'Avg_Open_To_Buy', 'Total_Trans_Amt', 'Total_Ct_Chng_Q4_Q1', 'Avg_Utilization_Ratio', 'Education_Level_Doctorate', 'Education_Level_Graduate', 
        'Education_Level_High School', 'Education_Level_Post-Graduate', 'Education_Level_Uneducated', 'Education_Level_Unknown', 
        'Income_Category_$40K - $60K', 'Income_Category_$60K - $80K', 'Income_Category_$80K - $120K', 'Income_Category_Less than $40K', 
        'Income_Category_Unknown', 'Marital_Status_Married', 'Marital_Status_Single', 'Marital_Status_Unknown', 'Card_Category_Others', 
        'Card_Category_Silver']]
y = df['Attrition_Flag']

sc = StandardScaler()
x = sc.fit_transform(x)

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=2025)

kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=2025)

lr = LogisticRegression(random_state=2025)
results = cross_validate(lr, x_train, y_train, cv=kfold, scoring=['accuracy', 'precision', 'recall', 'f1'])
print('######### Logistic Regression #########')
print('LR Accuracy:', results['test_accuracy'].mean())
print('LR Precision:', results['test_precision'].mean())
print('LR Recall:', results['test_recall'].mean())
print('LR F1 Score:', results['test_f1'].mean())

svc = SVC(probability=True, random_state=2025)
results = cross_validate(svc, x_train, y_train, cv=kfold, scoring=['accuracy', 'precision', 'recall', 'f1'])
print('\n######### Support Vector Machine #########')
print('SVC Accuracy:', results['test_accuracy'].mean())
print('SVC Precision:', results['test_precision'].mean())
print('SVC Recall:', results['test_recall'].mean())
print('SVC F1 Score:', results['test_f1'].mean())

knc = KNeighborsClassifier()
results = cross_validate(knc, x_train, y_train, cv=kfold, scoring=['accuracy', 'precision', 'recall', 'f1'])
print('\n########## KNeighbors Classifier ##########')
print('KNC Accuracy:', results['test_accuracy'].mean())
print('KNC Precision:', results['test_precision'].mean())
print('KNC Recall:', results['test_recall'].mean())
print('KNC F1 Score:', results['test_f1'].mean())

clf_DT = tree.DecisionTreeClassifier(random_state=2025)
results = cross_validate(clf_DT, x_train, y_train, cv=kfold, scoring=['accuracy', 'precision', 'recall', 'f1'])
print('\n########## Decision Tree ###########')
print('DT Accuracy:', results['test_accuracy'].mean())
print('DT Precision:', results['test_precision'].mean())
print('DT Recall:', results['test_recall'].mean())
print('DT F1 Score:', results['test_f1'].mean())

rf_clf = RandomForestClassifier(n_estimators=100, random_state=2025)
results = cross_validate(rf_clf, x_train, y_train, cv=kfold, scoring=['accuracy', 'precision', 'recall', 'f1'])
print('\n########## Random Forest ##########')
print('RF Accuracy:', results['test_accuracy'].mean())
print('RF Precision:', results['test_precision'].mean())
print('RF Recall:', results['test_recall'].mean())
print('RF F1 Score:', results['test_f1'].mean())

extt_clf = ExtraTreesClassifier(n_estimators=100, random_state=2025)
results = cross_validate(extt_clf, x_train, y_train, cv=kfold, scoring=['accuracy', 'precision', 'recall', 'f1'])
print('\n########## Extra Trees ###########')
print('ETC Accuracy:', results['test_accuracy'].mean())
print('ETC Precision:', results['test_precision'].mean())
print('ETC Recall:', results['test_recall'].mean())
print('ETC F1 Score:', results['test_f1'].mean())

bagg_clf = BaggingClassifier(estimator=clf_DT, n_estimators=100, random_state=2025)
results = cross_validate(bagg_clf, x_train, y_train, cv=kfold, scoring=['accuracy', 'precision', 'recall', 'f1'])
print('\n########## Bagging Classifier ##########')
print('BC Accuracy:', results['test_accuracy'].mean())
print('BC Precision:', results['test_precision'].mean())
print('BC Recall:', results['test_recall'].mean())
print('BC F1 Score:', results['test_f1'].mean())

adb_clf = AdaBoostClassifier(estimator=clf_DT, n_estimators=100, learning_rate=0.1, algorithm='SAMME', random_state=2025)
results = cross_validate(adb_clf, x_train, y_train, cv=kfold, scoring=['accuracy', 'precision', 'recall', 'f1'])
print('\n########## Ada Boost Classifier ##########')
print('ABC Accuracy:', results['test_accuracy'].mean())
print('ABC Precision:', results['test_precision'].mean())
print('ABC Recall:', results['test_recall'].mean())
print('ABC F1 Score:', results['test_f1'].mean())

gb_clf = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, random_state=2025)
results = cross_validate(gb_clf, x_train, y_train, cv=kfold, scoring=['accuracy', 'precision', 'recall', 'f1'])
print('\n######### Gradient Boosting ###########')
print('GBC Accuracy:', results['test_accuracy'].mean())
print('GBC Precision:', results['test_precision'].mean())
print('GBC Recall:', results['test_recall'].mean())
print('GBC F1 Score:', results['test_f1'].mean())

xgb_clf = xgb.XGBClassifier(n_estimators=100, objective='binary:logistic', seed=2025)
xgb_clf.fit(x_train, y_train, eval_set=[(x_test, y_test)], verbose=True)
print("""We can see the model was learning well till 40, after that it starts to increase
again, so we can use Early Stopping Rounds around 10-15, so it will stop earlier!""")
results = cross_validate(xgb_clf, x_train, y_train, cv=kfold, scoring=['accuracy', 'precision', 'recall', 'f1'])
print('\n########## XGBoost ##########')
print('XGB Accuracy:', results['test_accuracy'].mean())
print('GBC Precision:', results['test_precision'].mean())
print('GBC Recall:', results['test_recall'].mean())
print('GBC F1 Score:', results['test_f1'].mean())
```
